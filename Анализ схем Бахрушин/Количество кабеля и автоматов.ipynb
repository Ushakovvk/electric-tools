Группировка оборудования

```python
def full_desc(row):
    if row[3] == 3:
        Un = '~380 В'
    else:
        Un = '~220 В'
    text = f'{row[0]} {row[1]}, {row[2]:.0f} кВт, {Un}'
    return text
```

```python
data.loc[(data['Модель'].isna()) & (data['Код оборудования'].notna()), 'Модель'] = 'Оборудование без модели'
```

```python
#data = data.sort_values(['Код оборудования'])
equip_group = data.sort_values(['Код оборудования']).groupby(['Модель','Наименование','Pуст, кВт','Число фаз'])['Код оборудования'].agg([lambda eqs: ','.join(eqs), len])
equip_group.reset_index(inplace=True)
equip_group = equip_group.rename(columns={'<lambda_0>':'Позиция', 'len':'Количество'})
equip_group = equip_group.sort_values(by=['Позиция','Модель'])
equip_group['Модель'] = equip_group['Модель'].str.replace('Оборудование без модели', '')
equip_group['Полное наименование'] = equip_group[['Наименование','Модель', 'Pуст, кВт', 'Число фаз']].apply(full_desc, axis=1)
equip_group[['Позиция','Полное наименование','Количество']].to_excel('Перечень оборудования.xlsx')
equip_group
```

Группировка автоматов

```python
def parse_qf(t):
    return int(t[2:])
def join_qf(text):
    text_list = text.split(',')
    new_text = []
    
    if 'QF0' in text_list:
        text_list.remove('QF0')
        new_text.append('QF0')
        
    if text_list and (len(text_list) <= 2):
        new_text.extend(text_list)
        
    elif len(text_list) >= 3:
        qf_start = text_list[0]
        qf_end = text_list[0]
        for qf in text_list[1:]:
            if  (parse_qf(qf_end) + 1) == parse_qf(qf):
                qf_end = qf
            else:
                if qf_start == qf_end:
                    new_text.append(qf_start)
                    qf_start = qf
                    qf_end = qf
                else:
                    if parse_qf(qf_end) == (parse_qf(qf_start) + 1):
                        new_text.append(f'{qf_start},{qf_end}')
                    else:
                        new_text.append(f'{qf_start}-{qf_end}')
                    qf_start = qf
                    qf_end = qf
                    
        if qf_start == qf_end:
            new_text.append(qf_start)
        else:
            if parse_qf(qf_end) == (parse_qf(qf_start) + 1):
                new_text.append(f'{qf_start},{qf_end}')
            else:
                new_text.append(f'{qf_start}-{qf_end}')
    return ','.join(new_text)
```

```python
av_group = data.groupby(['Код щитка','Марка АВ'], as_index=True)
```

```python
names_av = av_group['Номер АВ'].agg([lambda eqs: ','.join(f'QF{int(x)}' for x in eqs), len])
names_av.reset_index(inplace=True)
names_av = names_av.rename(columns={'<lambda_0>':'Позиция', 'len':'Количество'})
names_av['Позиция'] = names_av['Позиция'].apply(join_qf)
names_av = names_av.sort_values(by=['Код щитка','Позиция'])
names_av['Позиция'] = names_av['Позиция'].str.replace('QF0','QFв')
names_av.loc[names_av['Марка АВ'].str.contains('Vigi'),'Марка АВ'] ='Дифференциальный автоматический выключатель ' + names_av[names_av['Марка АВ'].str.contains('Vigi')]['Марка АВ']
names_av.loc[names_av['Марка АВ'].str.contains('iC60N'),'Марка АВ'] ='Автоматический выключатель ' + names_av[names_av['Марка АВ'].str.contains('iC60N')]['Марка АВ']
names_av.loc[names_av['Марка АВ'].str.contains('ВА57'),'Марка АВ'] ='Автоматический выключатель ' + names_av[names_av['Марка АВ'].str.contains('ВА57')]['Марка АВ']
names_av.loc[names_av['Марка АВ'].str.contains('RBK'),'Марка АВ'] ='Выключатель-предохранитель ' + names_av[names_av['Марка АВ'].str.contains('RBK')]['Марка АВ']
names_av.to_excel('Автоматы щитков.xlsx')
names_av.head()
```
